1. 掷3个骰子1000次，各点（骰子点数之和）频率直方图

    - 使用面向对象编程创建骰子类；
    
    - 使用面向对象编程实现掷骰子；
    
    - 使用Matplotlib画各点频率直方图；

```python
# 使用面向对象编程创建骰子类并实现投掷。

class Dice:
    def __init__(self):
        # 骰子点数（1～6）
        self.num: int = 0
        
    def toss(self):
        self.num = random.randint(1, 6)
```

```python
# 投掷3个骰子各1000次，并用 matplotlib 绘制各点数频率的直方图。

DICES_COUNT = 3

dices = []
for _ in range(DICES_COUNT):
    dices.append(Dice())

# 投掷骰子1000次
data = []
for _ in range(1000):
    # 投掷3个骰子
    for d in dices:
        d.toss()
    # 累加并记录总共点数
    total = 0
    for d in dices:
        total += d.num
    data.append(total)

# 统计并绘制点数数据的直方图（条形图）
labels = np.arange(1, DICES_COUNT * 6 + 1)
nums = np.zeros(DICES_COUNT * 6)
for label in labels:
    for d in data:
        if label == d:
            nums[label - 1] += 1
plt.bar(labels, nums)
plt.show()
```

2. 利用numpy中的random画4个分布的频率直方图

    - 熟练使用numpy中的random模块产生随机数；
    
    - 产生随机数符合以下4种分布：随机分布、正态分布、二项分布、泊松分布；
    
    - 使用Matplotlib画4个分布图的频率直方图；

```python
# 使用 random 模块产生随机数
print(random.random())
```

```python
# 产生符合均匀分布的数据
COUNT = 10000
data = np.random.uniform(low=0.0, high=1.0, size=COUNT)
plt.hist(data, range=(0.0, 1.0))
plt.show()
```

```python
# 产生符合二项分布的数据
COUNT = 10000
N = 100
data = np.random.binomial(n=N, p=0.5, size=COUNT)
labels = np.arange(N)
nums = np.zeros(N)
for d in data:
    for n in range(N):
        if n == d:
            nums[n] += 1
plt.bar(labels, nums)
plt.show()
```

```python
# 产生符合泊松分布的数据
COUNT = 10000
L = 2.0
DISPLAY_COUNT = 20
data = np.random.poisson(lam=L, size=COUNT)
labels = np.arange(DISPLAY_COUNT)
nums = np.zeros(DISPLAY_COUNT)
for d in data:
    for n in range(DISPLAY_COUNT):
        if n == d:
            nums[n] += 1
plt.bar(labels, nums)
plt.show()
```

```python
# 产生符合正态分布（标准正态分布）的数据
COUNT = 10000
data = np.random.normal(loc=0.0, scale=1.0, size=COUNT)
plt.hist(data, range=(-5.0, 5.0), bins=100)
plt.show()
```
